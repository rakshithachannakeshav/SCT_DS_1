```python
import pandas as pd

# Adjust filename to your CSV file
file = "API_SP.POP.TOTL_DS2_en_csv_v2_571486.csv"

# Skip the first 4 metadata rows
df = pd.read_csv(file, skiprows=4)

# Preview
print(df.head())
```

```python
# Select country + population for 2022
df_pop = df[["Country Name", "Country Code", "2022"]].copy()
df_pop.columns = ["country", "code", "population"]

# Drop NaN
df_pop = df_pop.dropna(subset=["population"])

# Convert to millions for readability
df_pop["population_millions"] = df_pop["population"] / 1_000_000

print(df_pop.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist(df_pop["population_millions"], bins=50)
plt.xlabel("Population (millions)")
plt.ylabel("Number of countries")
plt.title("Distribution of Country Populations (2022)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df_pop["population_millions"], bins=50)
plt.xscale("log")
plt.xlabel("Population (millions, log scale)")
plt.ylabel("Number of countries")
plt.title("Distribution of Country Populations (2022) [log scale]")
plt.show()
```

```python
# Sort by population
top20 = df_pop.sort_values("population", ascending=False).head(20)

plt.figure(figsize=(10,8))
plt.barh(top20["country"], top20["population_millions"])
plt.xlabel("Population (millions)")
plt.title("Top 20 Most Populous Countries (2022)")
plt.gca().invert_yaxis()  # largest at top
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load CSV (skip metadata rows)
df = pd.read_csv(r"API_SP.POP.TOTL_DS2_en_csv_v2_571486.csv", skiprows=4)

# 2. Keep only India + Population total
india = df[(df["Country Name"] == "India") & (df["Indicator Name"] == "Population, total")]

# 3. Drop non-year columns (Country, Code, Indicator, and trailing NaN column)
india = india.drop(columns=["Country Name","Country Code","Indicator Name","Indicator Code","Unnamed: 69"])

# 4. Melt year columns into rows
india_years = india.melt(var_name="Year", value_name="Population")

# 5. Clean Year  convert to integer
india_years["Year"] = india_years["Year"].astype(int)

# 6. Plot Indias population trend
plt.figure(figsize=(10,6))
plt.plot(india_years["Year"], india_years["Population"], marker="o", color="teal")
plt.title("India - Population, total (19602024)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.grid(True)
plt.show()
```

```python
countries = ["India", "China", "United States"]
subset = df[(df["Country Name"].isin(countries)) & (df["Indicator Name"]=="Population, total")]
subset = subset.drop(columns=["Country Code","Indicator Name","Indicator Code","Unnamed: 69"])
subset = subset.melt(id_vars="Country Name", var_name="Year", value_name="Population")
subset["Year"] = subset["Year"].astype(int)

plt.figure(figsize=(12,6))
for country in countries:
    data = subset[subset["Country Name"]==country]
    plt.plot(data["Year"], data["Population"], marker="o", label=country)

plt.title("Population Growth (19602024)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = india_years["Year"].values.reshape(-1,1)
y = india_years["Population"].values

model = LinearRegression()
model.fit(X, y)

future_years = np.arange(2025, 2036).reshape(-1,1)
future_pop = model.predict(future_years)

plt.figure(figsize=(10,6))
plt.plot(india_years["Year"], india_years["Population"], label="Historical", marker="o")
plt.plot(future_years, future_pop, label="Forecast", marker="x", linestyle="--", color="red")
plt.title("India Population Forecast (20252035)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()
plt.grid(True)
plt.show()
```
